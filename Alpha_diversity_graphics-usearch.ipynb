## Functions

```python
# starting to write functions so I can actually re-use and read this code later.
# adjust_year fixes the year string to reflect whether samples are from 2014 or 2015.
def adjust_year(date):
    """ Replace the year with '15 if sample was taken in October thru December.
    Takes a date string formatted as mm/dd/yy and returns the same"""
    year =  int(date.split("/")[2])
    month = int(date.split("/")[0])
    if month >7:
        year="14"
    else: 
        year ="15"
    return str(month)+"/"+date.split("/")[1]+"/"+str(year)

# this imports a collated alpha diversity file and puts the highest rarefaction point in a dataframe.
def read_collated_alpha(data, row):
    metric = pd.read_csv(data, sep = "\t")
    length = np.shape(metric)[0]
    single_rare = metric.ix[length - row:, 3:]
    single_rare_float = single_rare.astype("float")
    return single_rare_float

def add_row(df, mapdf, rowID):
    row = mapdf[mapdf.SampleID==rowID].index
    date = mapdf.ix[row, "Datefixed"].values[0]
    date = adjust_year(date)
    new_date = datetime.datetime.strptime(date, '%m/%d/%y') # Assumes this format for all dates. 
    basin = mapdf.ix[row, "Reactor"].values[0]
    month = mapdf.ix[row, "Month"].values[0]
    description = mapdf.ix[row, "Description"].values[0]
    df.loc[rowID] = [new_date, basin, month, description]
    return df

def group_df(df, group):
# Sort dataframe into different groups 
    grouped  = df.groupby(group)
    keys = grouped.groups.keys()
    return grouped, keys

def get_season(month):
    seasons = {'January': 'Winter',
               'February': 'Winter',
                'March': 'Spring',
                'April': 'Spring',
                'May': 'Spring',
                'June': 'Summer',
                'July': 'Summer',
                'August': 'Summer',
                'September': 'Fall',
                'October': 'Fall',
                'November': 'Fall',
                'December': 'Winter'
                }
    return seasons[month]

#print get_season('July')
#print read_collated_alpha("/Users/jimbo/Desktop/6.9.15_seq_data/alpha_diversity/alpha_rare_uparse/alpha_collated3/simpson_reciprocal.txt",10)
```
