#### completed functions

```python
# 4. 
def abundance_sort(df, ID_col):
    # Takes dataframe of uparse OTUs with size in the OTU name and returns a sorted list of OTUs and total counts.
    otu_labels =df.ix[:,ID_col].values
    split_labels = [(x, x.split(";size=")[1]) for x in otu_labels]
    split_labels = [(x[0], float(x[1].strip(";"))) for x in split_labels]
    split_labels.sort(key=itemgetter(1), reverse=True)
    return split_labels
# return split_labels or x[1] for x in split_labels


# 5.
def frequency_sort(df, ID_col):
    # Takes dataframe of OTUs and returns a sorted list of OTUs by frequency of appearance.
    length = len(df.ix[1,:])
    freq_tuples = []
    for i,row in df.iterrows():
        counts = row.value_counts()
        zeros = 0
        try:
            zeros=counts.loc[0]
        except KeyError:
            pass
        frequency = length - zeros
        freq = (row.ix[ID_col], frequency)
        freq_tuples.append(freq)
    freq_tuples.sort(key=itemgetter(1), reverse=True)
    return freq_tuples


#6. Plotting
def run_sum(values):
    # Takes a vector, returns the running sum at each position in the vector. 
    running_sum = [0]
    cum_sum = sum(values)
    for i in values:
        next_val = running_sum[-1]+i/cum_sum
        running_sum.append(next_val)
    return running_sum

def plot_sorted_and_cum(values,running_sum):
    # Plots a vector of values and the cumulative suma at each value.
    fig,ax1 = plt.subplots()
    ax2 = plt.twinx(ax1)
    ax1.plot(values)
    ax2.plot(running_sum)
    return fig, ax1, ax2


# 7. Filter
# I want two filters - a. Keep value > x in [(name, x)...] or keep 0-sum(x) > something for same tuple_list.
def cumulative_threshold(tuplelist,threshold):
    # take a sorted list of tuples, return sorted list of tuples up to a certain % of total sequences kept.
    values = [float(x[1]) for x in tuplelist]
    summed = run_sum(values)
    retained_values = [x for x in summed if x < threshold]
    indices = len(retained_values)
    rows_to_keep = [x[0] for x in tuplelist[:indices]]
    return rows_to_keep

def threshold_filter(tuplelist, threshold):
    # Take a sorted list of tuples and return those with frequency > threshold
    values = [float(x[1]) for x in tuplelist]
    retained_values = [x for x in values if x >= threshold]
    rows_to_keep = [x[0] for x in tuplelist[:len(retained_values)]]
    return rows_to_keep

def filter_rows(df, col_to_filter, row_list):
    # filter - use the following syntax ---
    filtered_df = df[df.ix[:,col_to_filter].isin(row_list)]
    return filtered_df

#print cumulative_threshold(split_labels,.2)
#print threshold_filter(freq_tuples,156)
```

### Plotting raw data to see where to rarefy to

```python
# 6. plotting cumulative OTU data
# Plot abundance data 
split_labels = abundance_sort(otu_table, "OTUId")
log_ordered = [math.log10(x[1]) for x in split_labels]
relative_abundances = [x[1] for x in split_labels]
summe = run_sum(relative_abundances)
fig, ax1, ax2 = plot_sorted_and_cum(log_ordered, summe)

ax1.set_xlabel("OTU Rank")
ax1.set_ylabel("Log10(abundance)")
ax2.set_ylabel("Cumulative fraction of sequences included")
ax2.set_ylim([0,1])
plt.show()


# Plotting frequency data
#freq_tuples = frequency_sort(otu_table, "OTUId")
size = [(x[0].split("size=")[1].strip(";"),x[1]) for x in freq_tuples]
freqs = [x[1] for x in size]
sizes = [float(x[0]) for x in size]
summed = run_sum(sizes)

fig, ax1, ax2 = plot_sorted_and_cum(freqs, summed)
ax1.set_xlabel("OTU Rank")
ax1.set_ylabel("Samples present in")
ax2.set_ylabel("Cumulative fraction of sequences included")
plt.show()

```
