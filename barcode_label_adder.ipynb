```python
# Reverse complement barcodes in mapping file.

def ReverseComplement1(seq):
    seq_dict = {'A':'T','T':'A','G':'C','C':'G'}
    return "".join([seq_dict[base] for base in reversed(seq)])


barcodes = map_df["BarcodeSequence"].values
print barcodes
new_barcodes = []
for code in barcodes:
    new_code = ReverseComplement1(code)
    new_barcodes.append(new_code)
print new_barcodes

map_df["new_barcodes"] = new_barcodes
print map_df

map_df.to_csv("/Users/jimbo/Desktop/nanxi/CZO_map_reversed_barcodes.txt", sep="\t")
```
