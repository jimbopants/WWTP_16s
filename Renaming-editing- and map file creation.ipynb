```python
# This cell adds the dates in two formats mm/dd/yy, 
# and Month to an existing map file with the date in mm/dd format

import pandas as pd

# Import map file, add Month and New Date column
df = pd.read_excel("../map/v1/map_v1_corrected.xlsx")
df["Month"] = pd.Series(" 1", index=df.index)
df["Datefixed"] = pd.Series(" 1", index=df.index)
df['Date'] = df['Date'].astype(str)

dfreact = df.Reactor
dfdate = df.Date

# I want to split the date, check month, and then concat the date + 14 or 15 to a new column
# then based on the month (split[0]) I want to add a new value to the month column based on a dict of months
# Then I want to skip the RAR and NaN samples





# Creating month dict
months = {'1': "January",
          '2': "February",
          '3': "March",
          '4': "April",
          '5': "May",
          '8': 'August',
          '10': "October",
          '11': "November",
          '12': "December",
          }
#print months

# Adding month as a categorical value
for date in dfdate[:-1]: # Ignore the NaN value for the NTC.
        if "/" in date:
            x=date.split("/")
            row = df[df.Date==date].index # Super hacky way of getting index of a column value match.
            df.ix[row, "Month"] = months[x[0]]
            if int(x[0]) >5:
                df.ix[row, "Datefixed"] = x[0]+"/"+x[1]+"/"+"14"
            elif int(x[0]) <5:
                df.ix[row, "Datefixed"] = x[0]+"/"+x[1]+"/"+"15"
    
print df

df.to_excel("../map/map_v2.xlsx")
    
```
